### Hierarchical Clustering

```python
df1 = df.copy()

from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df1['Place'] = le.fit_transform(df1['Place'])
```

```python
df1
```

```python
df1.isna().sum()
```

```python
# Function for Normalizing the data

def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df1_norm = df1.iloc[:,1:].apply(norm_func)
df1_norm
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering

# Dendrogram

dendrogram = sch.dendrogram(sch.linkage(df1_norm, method='average'))
```

```python
hc = AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
hc
```

```python
y_hc = hc.fit_predict(df1_norm)
y_hc
```

```python
df1['Clusters'] = hc.labels_
df1
```

### K-Means Clustering 

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

df2 = df1.drop(columns='Clusters')
df2
```

```python
wcss=[]

for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df1_norm)
    wcss.append(kmeans.inertia_)

plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of cluster')
plt.ylabel('WCSS')
```

```python
kmc = KMeans(n_clusters=4)
kmc.fit(df1_norm)
kmc.labels_
```

```python
df2['Clusters'] = kmc.labels_
df2
```

```python
df2.iloc[:,1:5].groupby(df2.Clusters).mean()
```

### DBSCAN Clustering

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

df3 = df1.drop(columns='Clusters')
df3
```

```python
sc = StandardScaler()
df3_scaled = sc.fit_transform(df3.iloc[:,1:5])
df3_scaled
```

```python
dbscan = DBSCAN(eps=2,min_samples=5)
dbscan.fit(df3_scaled)
```

```python
df3['Clusters'] = dbscan.labels_
df3
```
