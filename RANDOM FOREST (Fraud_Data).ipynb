```python
# Importing Necessary Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.figsize'] = (10,5)
plt.rcParams['figure.dpi'] = 300
```

```python
df = pd.read_csv("D:\ExcelR\Data Science\Assignments\Decission Trees\Fraud_check.csv")
df
```

```python
df['Undergrad'] = df['Undergrad'].map({'YES':'1', 'NO':'0'}).astype(int)
df['Marital.Status'] = df['Marital.Status'].map({'Single':'0', 'Married':'1', 'Divorced':'2'}).astype(int)
df['Urban'] = df['Urban'].map({'YES':'1', 'NO':'0'}).astype(int)

df['Taxable.Income'] = df['Taxable.Income'].apply(lambda x: '0' if x <= 30000 else '1')
df
```

```python
# Defining 'Features' (x) & 'Target' (y) variables

x = df.drop(columns='Taxable.Income')
y = df['Taxable.Income']

from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.2, random_state=42)
```

### Model Building

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

rf = RandomForestClassifier(n_estimators=15, criterion='entropy')
rf.fit(xtrain, ytrain)
rf_pred = rf.predict(xtest)

print(f"Training Accuracy ;", rf.score(xtrain, ytrain))
print(f"Testing Accuracy ;", rf.score(xtest, ytest))
print(f"Model Accuracy ;", classification_report(ytest, rf_pred))
```
